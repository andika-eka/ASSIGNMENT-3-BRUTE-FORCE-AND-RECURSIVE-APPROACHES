```python
def printTower(disk, t1, t2, t3):
    towers = list()
    towers.append(t1)
    towers.append(t2)
    towers.append(t3)

    print("moving :",disk )

    for tower in towers:
        if tower[1] == "source":
            print(tower[1], " : ", tower[0])
            break
    for tower in towers:
        if tower[1] == "helper":
            print(tower[1], " : ", tower[0])
            break
    for tower in towers:
        if tower[1] == "target":
            print(tower[1], " : ", tower[0])
            break

def hanoi(n, source, helper, target):
    if n > 0:
        hanoi(n - 1, source, target, helper)
        if source[0]:
            disk = source[0].pop()
            target[0].append(disk)
            printTower(n, source, helper, target)
        hanoi(n - 1, helper, source, target)
        
source = ([5,4,3,2,1], "source")
target = ([], "target")
helper = ([], "helper")

printTower(0,source,helper,target)
hanoi(len(source[0]),source,helper,target)
printTower(0,source,helper,target)
```
